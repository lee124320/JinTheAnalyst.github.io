```python
import random


#initialize variables and read in word files

wins = 0
losses = 0
used_words = []
max_guesses = 7

with open("words_file.txt", "r") as words_file:
    words = words_file.read()
    global words_list   #global bc we're defining a variable and want it to be accessible throughout the whole program
    words_list = list(map(str, words.split('\n')))

def display_welcome_message():
    print('Welcome to The Word Guessing Game!')
    print("A random word will be chosen from a list and you will try to guess what it is.")
    print(f'You will have a max of {max_guesses} incorrect guesses.')
    print('Good Luck!')
    
display_welcome_message()

def pick_word(words_list):
    word = random.choice(words_list)
    print('A word has been chosen!')
    print("_ " * len(word))
    return word

while True:
    word = pick_word(words_list)
    if word in used_words:
        print(f'You already guessed {word}... I will pick another word')
        continue
    used_words.append(word)
    guess_count = 0
    display_word = "_ " * len(word)
    guesses = []
    
    #ToDo = Fix infinite loops
    while True:
        guess = input('Please guess a letter.')
        # data validation - must be a letter
        if guess.isalpha()is False:
            print('Please enter a letter or word.')
            continue
        # no penalty for re-guesses
        if guess in guesses:
            print('You have already guessed this. Please try again.')
            continue
        #append guess to guesses
        guesses.append(guess)
        
        if guess in word:
            #build list of positions of guessed letter in the word
            #startswith: string.startswith(value, start, end) returns true if the string starts with the specified value
            position = [i for i in range(len(word)) if word.startswith(guess, i)]
            for i in position:
                display_word = display_word[:i] + guess + display_word[i+ len(guess):]
                print(f'The Word:{display_word}')
            if "_" not in display_word:
                print('All Letters Have Been Guessed Correctly, You Win!')
                wins += 1
                break
        
        else:
            guess_count +=1
            if guess_count == max_guesses:
                print(f'Too many tries. Sorry, you lose. The word was {word}.')
                losses +=1
                break
                print(f'You have {max_guesses - guess_count} guesses left')
        
    #play again?
    while True:
        global play_again
        play_again = input(f'Win Count: {wins} wins. {losses} losses.\nPlay again? (y/n)')
        if (play_again != 'y'and play_again != 'n'):
            print('Invalid response. Please select y or n')
        else:
            break
    if play_again == 'n': 
        break
                      
    print('Goodbye. Thanks for playing.')
            
```
