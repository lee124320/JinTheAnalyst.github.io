```python
sns.set_style("white")
list1 = []
mylabels = []
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == gen].BudgetMillions)
    mylabels.append(gen)

sns.set_style("whitegrid")  
# to adjust size
fig, ax = plt.subplots()  
fig.set_size_inches(11.7, 8.26) # size of A4 paper

h1 = plt.hist(list1, bins=30, stacked=True, label=mylabels)

plt.title("Movie Buget Distribution", fontsize=35, \
    color = 'DarkBlue')
plt.ylabel ("Number of Movies")
plt.xlabel("Budget")
plt.xticks(fontsize=20)

plt.legend(frameon=True, fancybox=True, shadow=True, framealpha=1, prop={'size':20})
plt.show()
```
